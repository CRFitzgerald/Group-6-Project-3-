```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.goodreads.com/list/show/19.Best_for_Book_Clubs'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
books = []

for book in soup.find_all('tr', {'itemtype': 'http://schema.org/Book'}):
    book_details = {}
    
    # Get the book cover image URL
    book_details['image_url'] = book.find('img', {'class': 'bookCover'}).get('src')
    
    # Get the book title and author
    book_details['title'] = book.find('a', {'class': 'bookTitle'}).text.strip()
    book_details['author'] = book.find('a', {'class': 'authorName'}).text.strip()
    
    # Get the book rating, score and number of votes
    rating = book.find('span', {'class': 'minirating'})
    if rating:
        rating_text = rating.text.strip()
        book_details['rating'] = rating_text.split()[0]
        book_details['score'] = rating_text.split()[2]
        book_details['votes'] = rating_text.split()[4][1:-1]
    else:
        book_details['rating'] = None
        book_details['score'] = None
        book_details['votes'] = None
    
    books.append(book_details)
```

```python
for book in books:
    print(book['title'], 'by', book['author'], 'has a rating of', book['rating'], 'with', book['votes'], 'votes.')
```

```python
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd
```

```python
url = "https://www.goodreads.com/list/show/19.Best_for_Book_Clubs"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
books = []
for book in soup.find_all('tr', {'itemtype': 'http://schema.org/Book'}):
    title = book.find('a', {'class': 'bookTitle'}).text.strip()
    author = book.find('a', {'class': 'authorName'}).text.strip()
    rating = book.find('span', {'class': 'minirating'}).text.strip()
    score_element = book.find('span', text='score:')
    score = score_element.next_sibling.strip() if score_element else None
    votes_element = book.find('span', text='people voted')
    votes = votes_element.previous_sibling.strip() if votes_element else None
    img_url = book.find('img')['src']
    books.append({'title': title, 'author': author, 'rating': rating, 'score': score, 'votes': votes, 'img_url': img_url})
```
